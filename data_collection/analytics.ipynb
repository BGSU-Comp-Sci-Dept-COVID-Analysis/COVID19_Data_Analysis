```python
all_data = pd.DataFrame()
for f in glob.glob("CovidTrackingProject/Daily/State/*/*.csv"):
    df = pd.read_csv(f)
    all_data = all_data.append(df,ignore_index=True)
```

```python
all_data['date'] = pd.to_datetime(all_data['date'], format='%Y%m%d')
```
